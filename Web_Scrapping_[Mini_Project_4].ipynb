### Def Version

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

def scrape_kompas():
    alamat = "https://www.kompas.com/"
    html = urlopen(alamat)
    data = BeautifulSoup(html, 'html.parser')

    table = data.findAll("div", {"class": "mostWrap"})[0]
    rows = table.findAll("a")

    hasil = []
    for row in rows:
        info = []
        for cell in row.findAll(["h2"]):
            info.append(cell.get_text())
        hasil.append(info)

    df = pd.DataFrame(hasil, columns=['Judul Berita Terpopuler'])
    df = df.replace('\n', '', regex=True)
    df["Judul Berita Terpopuler"] = df["Judul Berita Terpopuler"].str.strip()
    df = df.drop([0, 6])

    print(df)

# Call the function to scrape Kompas.com and print the resulting dataframe
scrape_kompas()
```
