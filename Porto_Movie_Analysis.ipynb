# Data Preparation

### Data Type Conversion


```python
df["release_date"] = pd.to_datetime(df["release_date"], errors='coerce')
df["budget"] = pd.to_numeric(df["budget"], errors='coerce')
```

### Drop Duplicates Data

Delete duplicate data based on original title, title, release date, and imdb id

```python
check_duplicates = ['original_title', 'release_date', 'imdb_id']

df[df.duplicated(subset=check_duplicates, keep=False)]
df = df.drop_duplicates(subset=check_duplicates)
```

### Extraction of Value "Name" in Dictionary

```python
from ast import literal_eval


features = ['genres', 'spoken_languages']
for feature in features:
  df[feature] = df[feature].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

# **The best voted Movie from the 2000s**

**Find the weighted_rating to determine the movie with the best vote based on the score:**

```python
m = best_movie_2000s['vote_count'].quantile(0.9)
C = best_movie_2000s['vote_average'].mean()

def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
best_movie_2000s['score'] = best_movie_2000s.apply(weighted_rating, axis=1)
```

```python
best_movie_2000s_by_vote = best_movie_2000s.sort_values(by='score', ascending=False).head(10)
best_movie_2000s_by_vote
```

```python
barbestmoviebyvote = best_movie_2000s_by_vote.groupby('title')['score'].sum().sort_values(ascending=False).head(10)
barbestmoviebyvote
```

> The analysis of movie ratings reveals that the top-rated films include "The Dark Knight" with a score of `8.24`, followed closely by "Whiplash" at `8.15` and "Spirited Away" at `8.13`. Other highly-rated films are "The Intouchables" `8.08`, "Inception" `8.05`, and "Interstellar" `8.04`. Additionally, "The Lord of the Rings: The Return of the King" scores `8.02`, while "Memento" achieves `7.95`, and "The Lord of the Rings: The Fellowship of the Ring" and "The Lord of the Rings: The Two Towers" score `7.93` and `7.92`, respectively.
